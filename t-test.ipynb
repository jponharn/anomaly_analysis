```python
import pandas as pd
from scipy.stats import ttest_ind
```

```python
result_1hr = pd.read_csv("cluster_data/result_1hr.csv")
result_1hr
```

```python
result_15min = pd.read_csv("cluster_data/result_15min.csv")
result_15min
```

```python
# result_1hr[result_1hr.station == result_15min.station]
int_df = pd.merge(result_1hr, result_15min, how ='inner', on =['station', 'DateTime'])
int_df
```

```python
ttest_ind(int_df.kmeans_umap_labels_x, int_df.kmeans_umap_labels_y)
```

```python
ttest_ind(result_1hr.kmeans_AE_labels, result_15min.kmeans_AE_labels)
```

```python
ttest_ind(result_1hr.kmeans_tsne_labels, result_15min.kmeans_tsne_labels)
```

# Graph

```python
from bokeh.plotting import figure, output_file, show
from bokeh.transform import dodge
```

```python
reduction_time = pd.DataFrame({
    "methods": ["Autoencoder", "t-SNE", "UMAP"],
    "5min": [533.23, 30.45, 82.99],
    "1hour": [24.08, 21.21, 65.98]
})
reduction_time
```

```python
reduction_time.methods
```

```python
fig = figure(x_range = reduction_time.methods, plot_width = 300, plot_height = 300, x_axis_label = "Dimensional Reduction Methods", y_axis_label ="Time (seconds.)")
fig.vbar(x = dodge('methods', -0.25, range = fig.x_range), top = '5min',
width = 0.2,source = reduction_time, legend_label="5min", color = "green")
fig.vbar(x = dodge('methods', 0.0, range = fig.x_range), top = '1hour',
width = 0.2, source = reduction_time,legend_label="1hour",color = "blue")
show(fig)
```
